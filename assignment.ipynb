```python
##Basics of Probability


#1. Write a Python program to simulate the following scenarios:  
 # a. Tossing a coin 10,000 times and calculating the experimental probability of heads and tails.  
import random

# a. Tossing a coin 10,000 times
def coin_toss_simulation(trials=10000):
    heads = 0
    for _ in range(trials):
        toss = random.choice(['H', 'T'])  # Use list, not set
        if toss == 'H':
            heads += 1
    tails = trials - heads
    print(f"Experimental Probability of Heads: {heads / trials}")
    print(f"Experimental Probability of Tails: {tails / trials}")

# b. Rolling two dice and computing the probability of getting a sum of 7
def dice_sum_simulation(trials=10000):
    count = 0
    for _ in range(trials):
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
        if die1 + die2 == 7:
            count += 1
    print(f"Experimental Probability of getting a sum of 7: {count / trials}")

# Run both simulations
coin_toss_simulation()
dice_sum_simulation()

```

```python
#2. Write a function to estimate the probability of getting at least one "6" in 10 rolls of a fair die.  
def least_six():
    try :
        count = 0
        for _ in range(1000):
            rolls = [random.randint(1, 6) for _ in range(10)]
            if 6 in rolls:
                count += 1
    except:
        print("An error occurred during the simulation.")
    return count / 1000
print(f"Estimated Probability of getting at least one '6' in 10 rolls: {least_six()}")            
```

```python
#Conditional Probability and Bayes' Theorem


#3. A bag contains 5 red, 7 green, and 8 blue balls. A ball is drawn randomly, its color noted, and it is put back into the bag. If this process is repeated 1000 times, write a Python program to estimate:  
 # a. The probability of drawing a red ball given that the previous ball was blue. 
colors = ['R']*5 + ['G']*7 + ['B']*8  # Bag of balls
prev_colors = []
current_colors = []

for _ in range(1000):
    first = random.choice(colors)  # First pick
    second = random.choice(colors)  # Second pick
    prev_colors.append(first)
    current_colors.append(second) 
    count_blue_then_red = sum(1 for p, c in zip(prev_colors, current_colors) if p == 'B' and c == 'R')
total_blue_before = prev_colors.count('B')
p_red_given_blue = count_blue_then_red / total_blue_before if total_blue_before > 0 else 0
print("\n4a. P(Red | Blue before):", p_red_given_blue)
 # b. Verify Bayes' theorem with the simulation results.  
total_red = current_colors.count('R')
total_blue = prev_colors.count('B')
total_trials = len(prev_colors)
p_b = total_blue / total_trials
p_r = total_red / total_trials
p_b_given_r = sum(1 for p, c in zip(prev_colors, current_colors) if c == 'R' and p == 'B') / total_red if total_red > 0 else 0
p_r_given_b = p_red_given_blue
bayes_rhs = (p_r_given_b * p_b) / p_r if p_r > 0 else 0

print("4b. Bayes Theorem Verification:")
print("P(B | R) from simulation:", p_b_given_r)
print("P(B | R) from Bayes' Theorem:", bayes_rhs)
```

```python
#4. Generate a sample of size 1000 from a discrete random variable with the following distribution:  
import numpy as np
values = [1, 2, 3]
prob = [0.25, 0.35, 0.40]
sample = np.random.choice(values, size=1000, p=prob)  # Generate 1000 samples
print("\n5. Discrete Variable:")
print("Mean:", np.mean(sample))
print("Variance:", np.var(sample))
print("Std Dev:", np.std(sample))
```

```python
# 5. Simulate 2000 random samples from an exponential distribution with a mean of 5. Visualize the distribution using:  
  # a. A histogram.  
import matplotlib.pyplot as plt
exp_data = np.random.exponential(scale=5, size=2000)
plt.hist(exp_data, bins=30, density=True, alpha=0.7, label='Exponential')
plt.title('Exponential Distribution (mean = 5)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# 6. Simulate the Central Limit Theorem
uniform_data = np.random.uniform(0, 1, 10000)
means = [np.mean(np.random.choice(uniform_data, size=30)) for _ in range(1000)]
plt.hist(means, bins=30, color='green', alpha=0.7)
plt.title('Central Limit Theorem - Sample Means (n=30)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```
