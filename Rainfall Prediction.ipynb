```python
#importing libraries 
import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("Austin_weather_final.csv")
```

```python
data
```

```python
X = data.drop(["PrecipitationSumInches"],axis = 1)
```

```python
Y = data["PrecipitationSumInches"]
```

```python
Y = Y.values.reshape(-1,1)
```

```python
Y
```

```python
day_index = 798
days = [i for i in range(Y.size)]
```

```python
days
```

```python
#initialise the linear regression model
lr = LinearRegression()

#train the data
lr.fit(X,Y)
```

```python
#plot a graph
print("The percipitation trend graph :")
plt.scatter(days, Y, color = 'g')
plt.scatter(days[day_index], Y[day_index], color = 'r')
plt.title("Percipitation Levels")
plt.xlabel("Days")
plt.ylabel("Percipitation in inches")
plt.show()
x_vis = X.filter(["TempAvgF" , "DewPointAvgF" , "HumidityAvgPercent", "SeaLevelPressureAvgInches", "VisibilityAvgMiles","WindAvgMPH	"]) 
```

```python
print("The precipitation trend graph :")

for i in range(x_vis.columns.size):
    plt.subplot(3, 2, i+1)
    
    # Green scatter for full column values
    plt.scatter(days[day_index], x_vis.iloc[day_index, i], color='g', label='Value at day_index')
    
    # Red scatter for the entire column trend
    plt.scatter(days, x_vis.iloc[:, i], color='r', s=10, label='Trend')
    
    plt.title(x_vis.columns[i])
    plt.legend()

plt.tight_layout()
plt.show()
```
